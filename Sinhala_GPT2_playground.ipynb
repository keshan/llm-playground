```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
```

```python
tokenizer = AutoTokenizer.from_pretrained("keshan/sinhala-gpt2")
model = AutoModelForCausalLM.from_pretrained('keshan/sinhala-gpt2-newswire')
```

```python
generator = pipeline('text-generation', model=model, tokenizer=tokenizer)
```

Play!!!!!!

```python
starting_text = "\u0D9A\u0DCA\u200D\u0DBB\u0DD2\u0D9A\u0DA7\u0DCA" #@param {type:"string"}
length_of_generated_text = 250 #@param {type:"number"}
num_examples = 5 #@param {type:"number"}

generator(starting_text, max_length=length_of_generated_text, num_return_sequences=num_examples)
```
